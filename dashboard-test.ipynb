# Example Dashboard

## 1. Bar Chart

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
valores = [10, 25, 15, 30]
plt.bar(labels, valores, color='skyblue')
plt.show()
```

## 2. Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(20)
y = np.random.rand(20)
plt.scatter(x, y, color='purple')
plt.show()
```

## 3. Line Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.show()
```
